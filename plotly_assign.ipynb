```python
#QUESTION 1
import seaborn as sns

# Load the Titanic dataset using seaborn
titanic_df = sns.load_dataset('titanic')
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic_df.age, y=titanic_df.fare, mode='markers' ))
fig.show()
```

```python
#QUESTION 2
import plotly.express as px
import seaborn as sns  # Only used to load the dataset
tips_df = sns.load_dataset('tips')
fig = px.box(tips_df, x='day', y='total_bill', title='Box Plot of Total Bill by Day',
             labels={'day': 'Day of Week', 'total_bill': 'Total Bill Amount'})

# Show the plot
fig.show()
```

```python
#QUESTION 3
import seaborn as sns
import plotly.express as px

# Load the tips dataset using seaborn
tips_df = sns.load_dataset('tips')

# Create the histogram using Plotly Express
fig = px.histogram(tips_df, x='sex', y='total_bill', color='day',
                   pattern_shape='smoker', title='Histogram of Total Bill by Sex',
                   labels={'sex': 'Sex', 'total_bill': 'Total Bill'})

# Show the plot
fig.show()
```
